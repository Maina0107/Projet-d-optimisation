Moyennes des temps pris pour résoudre les instances, lorsque l'instance est prouvée réalisable.

On dit prouvée réalisable, car ce n'est pas parce qu'on a pas trouvé de solution réalisable dans le temps imparti que le problème est irréalisable.\
On peut simplement avoir une instance avec trop de noeuds pour que le solveur trouve une solution réalisable dans le temps imparti, même si il en existe.

```python
for i in range (1,4):
    print("version ", i, " sans capacité : ", fwf["temps"][(fwf["statut"]==True) & (fwf["v"]==i) & (fwf["c"]==0)].mean())
```

```python
for i in range (1,4):
    print("version ", i, " avec capacité : ", fwf["temps"][(fwf["statut"]==True) & (fwf["v"]==i) & (fwf["c"]>0)].mean())
```
