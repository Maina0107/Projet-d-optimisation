Gap entre sol relax et sol entière

```python
def filter_and_update(fwf):
    # Grouper par les colonnes n, p, et i
    grouped = fwf.groupby(["n", "p", "i"])
    
    # Liste pour stocker les résultats
    rows = []
    
    for (n, p, i), group in grouped:
        if group["etat"].any():  # Vérifie si au moins une ligne a etat=True
            true_obj = group.loc[group["etat"], "obj"].iloc[0]  # Récupère obj où etat=True
            # Met à jour les lignes où etat=False
            group.loc[~group["etat"], "obj"] = true_obj
        rows.append(group)
    
    # Combine les groupes mis à jour
    return pd.concat(rows)
```

Sans capacités : 

```python
filter_fwf = filter_and_update(fwf[fwf["c"]==0])
```

```python
merged_df = pd.merge(
    filter_fwf[filter_fwf["obj"]!=-1].drop(columns=["temps", "temps_creation", "erreur", "gap", "obj_upper", "obj_lower", "validite"]),
    fwf_relax[(fwf_relax["etat"]==True) & (fwf_relax["c"]==0)].drop(columns=["temps", "temps_creation", "erreur", "gap", "obj_upper", "obj_lower"]),
    on=["n", "p", "i", "c", "v"],
    suffixes=("_df1", "_df2")
)
merged_df["ecart_pourcentage"] = round((abs(merged_df["obj_df2"] - merged_df["obj_df1"]) / merged_df["obj_df1"]) * 100, 2)

print(round(merged_df["ecart_pourcentage"].mean(), 2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Supposons que tes DataFrames soient déjà chargés dans des variables
dfs = {
    5: fwf_n5,
    10: fwf_n10,
    20: fwf_n20,
    30: fwf_n30,
    50: fwf_n50
}

# Dictionnaire pour stocker les moyennes des temps d'exécution et les comptages
results = {(v, c): [] for v in [1, 2, 3] for c in [1, 2]}
counts = {(v, c): [] for v in [1, 2, 3] for c in [1, 2]}
total_counts = {(v, c): [] for v in [1, 2, 3] for c in [1, 2]}

# Récupérer les valeurs de n
n_values = list(dfs.keys())

for n, df in dfs.items():
    for v in [1, 2, 3]:  # Pour chaque version
        for c in [1, 2]:  # Pour chaque capacité
            # On veut garder que les instances pour lesquelles on a l'optimalité
            df_filtered = df[(df["c"] == c) & (df["etat"] == True) & (df["v"] == v)]
            
            count = len(df_filtered)
            mean_time = df_filtered["temps"].mean() if count > 0 else 0
            
            results[(v, c)].append(mean_time)
            counts[(v, c)].append(count)
            
            # Compter le nombre total de lignes pour cette version et capacité
            total_count = 15  # Vous pouvez ajuster cette valeur si nécessaire
            total_counts[(v, c)].append(total_count)

# Tracé du graphique à barres groupées
plt.figure(figsize=(15, 8))

x = np.arange(len(n_values))  # Les positions des groupes de barres
width = 0.15  # La largeur des barres

# Définir les couleurs personnalisées pour chaque combinaison de v et c
colors = {
    (1, 1): '#2e34f0',   # beu foncé pour v=1, c=1
    (1, 2): '#4ECDC4',   # Turquoise pour v=1, c=2
    (2, 1): '#FFD93D',   # Jaune or pour v=2, c=1
    (2, 2): '#f59607',   # orange pour v=2, c=2
    (3, 1): '#8cf558',   # Vert d'eau pour v=3, c=1
    (3, 2): '#1a9909'    # vert foncé pour v=3, c=2
}

# Tracer les barres pour chaque version et capacité
for i, v in enumerate([1, 2, 3]):
    for j, c in enumerate([1, 2]):
        offset = (i*2 + j - 3) * width
        bars = plt.bar(x + offset, results[(v, c)], width, label=f'Version {v}, c={c}', color=colors[(v, c)])
        
        # Ajouter le nombre de lignes au-dessus de chaque barre
        for k, (bar, count, total) in enumerate(zip(bars, counts[(v, c)], total_counts[(v, c)])):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                     f'{count}/{total}',
                     ha='center', va='bottom', rotation=90)

plt.xlabel("Nombre de nœuds (n)")
plt.ylabel("Temps moyen d'exécution")
plt.title("Temps d'exécution moyen (en secondes) en fonction du nombre de nœuds, de la version et de la capacité")
plt.xticks(x, n_values)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
