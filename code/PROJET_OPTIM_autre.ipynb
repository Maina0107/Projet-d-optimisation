On compare maintenant la relaxation linéaire des deux formulations : on regarde le pourcentage d'écart moyen avec la solution optimale entière (lorsque les deux solutions sont optimales)

```python
# récupérer les données au début quand ce sera possible, puis supprimer cette cellule
fwf_relax = pd.DataFrame({
    "n": [5, 5, 5, 5, 5],
    "p": [1, 1, 1, 1, 5],
    "i": [1, 1, 2, 2, 3],
    "c": [1, 2, 1, 2, 1],
    "v": [1, 1, 1, 1, 1],
    "etat": [True, True, True, True, True],
    "obj": [50, 20, 80, 40, 80]
})
```

```python
merged_df = pd.merge(
    fwf[(fwf["etat"]==True) & (fwf["v"]==1) & (fwf["c"]>0)],
    fwf_relax[(fwf_relax["etat"]==True) & (fwf_relax["v"]==1) & (fwf_relax["c"]>0)],
    on=["n", "p", "i", "c", "v"],
    suffixes=("_df1", "_df2")
)
merged_df["ecart_pourcentage"] = round((abs(merged_df["obj_df2"] - merged_df["obj_df1"]) / merged_df["obj_df1"]) * 100, 2)
merged_df
```

```python
ecart_moyen_par_v = merged_df.groupby("c")["ecart_pourcentage"].mean().reset_index()
ecart_moyen_par_v.columns = ["c", "ecart_pourcentage_moyen"]
ecart_moyen_par_v
```

<h2 style="padding:16px; color:white; background:#07D">Données pour faire le graphique du temps mis en fonction du nb de noeuds</h2>

```python
data = fwf[["n", "v", "c", "statut", "temps_creation", "temps"]]
data.head()
```

Pour les instances irréalisables où on a pas trouvé de solution réalisable avant la fin du temps (donc pas n=5), on met le temps max.

```python
data.loc[(data["statut"]==False) & (data["n"]>5), "temps"] = 300
data = data[~((data["n"] == 5) & (data["statut"] == False))]
```

```python
data_ready = data.groupby(["v", "c", "n"]).agg(
    moyenne_temps=("temps", "mean"),
    moyenne_temps_creation=("temps_creation", "mean")
).reset_index()
data_ready
```
